# Visualise  the Categorical Variables

## OVERTIME

```python
sns.countplot( data = df, x = 'OverTime')
plt.title('OverTime')
plt.show()
```

## Marital Status

```python
sns.countplot(data = df, x ='MaritalStatus')
plt.title('MaritalStatus')
plt.show()
```

## Job Role

```python
sns.countplot(data = df, x = 'JobRole')
sns.set(rc={'figure.figsize':(20,10)})
plt.title('JobRole')
plt.show()
```

## Gender

```python
sns.countplot(data = df, x = 'Gender')
sns.set(rc={'figure.figsize':(2,4)})
plt.title('Gender')
plt.show()
```

## Education Field

```python
sns.countplot(data = df, y = 'EducationField')
sns.set(rc={'figure.figsize':(10,5)})
plt.title('EducationField')
plt.show()
```

## Department

```python
sns.countplot(data = df , y = 'Department')
sns.set(rc = {'figure.figsize':(20,10)})
plt.title('Department')
plt.show()
```

## Buisness Travel

```python
sns.countplot(data =df , x = 'BusinessTravel')
sns.set(rc = {'figure.figsize':(5,5)})
plt.title('BusinessTravel')
plt.show()
```

```python
# Relation between overtime and age

sns.boxplot(data = df , x = 'OverTime', y = 'Age')
plt.title('Relationship between Overtime and Age')
plt.show()
```

# Plotting Numerical Values

```python
## Total Working years

sns.histplot(data = df, x = 'TotalWorkingYears',bins = 10 , kde = True)
plt.title('TotalWorkingYears')
plt.show()
```

```python
# Education level

sns.histplot(data = df, x = 'Education', bins = 10,kde= True)
plt.title('Education Level')
plt.show()
```

```python
# Number of Companies Worked

sns.histplot(data = df, x = 'NumCompaniesWorked',bins = 10, kde = True)
plt.title('Number of Companies Worked')
plt.show()
```

```python
#Distance from home

sns.histplot(data = df, x =  'DistanceFromHome', bins = 10 , kde = True)
plt.title('Distance from home')
plt.show()
```

## Salary Analysis

```python
# Monthly income Distribution

sns.histplot(data = df, x = 'MonthlyIncome', bins = 10, kde = True)
plt.title('MonthlyIncome')
plt.show()


```

```python
# Salary by Job role

plt.figure(figsize = (20,10))
sns.boxplot(data = df, x = 'JobRole', y = 'MonthlyIncome')
plt.title('Salary by Job role')
plt.show()
```

# Employee Satisfaction

```python
# Job Satisfaction Distribution

sns.histplot(data = df, x = 'JobSatisfaction', bins = 10 , kde = True)
plt.title('Job Satisfaction Distribution')
plt.show()
```

```python
# Performance Rating Distribution

sns.histplot(data = df, x ='PerformanceRating', bins = 10, kde = True)
plt.title('Performance Rating Distribution')
plt.show()
```

```python
# Employee Tenure

sns.histplot(data = df, x = 'YearsAtCompany', bins = 10, kde= True)
plt.title('Employee Tenure ')
plt.show()
```

```python
# Age distribution by Department

plt.figure(figsize = (10,6))
sns.boxplot(data = df , x = 'Department', y = 'Age')
plt.title('Age distribution by Department')
plt.show()
```

```python
# Attrition by Job Role

plt.figure(figsize = (20,10))
sns.countplot(data= df , x = 'JobRole', hue = 'Attrition')
plt.title('Attrition by Job Role ')
plt.show()
```
